# Comparisons

```python
# Group by tournament_winner = "Yes" and "No", display adjusted_offensive_efficiency
tourney_winners = df[df['tournament_winner'] == 'Yes']['adjusted_offensive_efficiency']
non_tourney_winners = df[df['tournament_winner'] == 'No']['adjusted_offensive_efficiency']
print("Tournament Winners - Adjusted Offensive Efficiency:")
print(tourney_winners.describe())
print("\nNon-Tournament Winners - Adjusted Offensive Efficiency:")
print(non_tourney_winners.describe())
```

```python
# Create stacked bar plot for adjusted_offensive_efficiency by year for tournament winners vs non-tournament winners
plt.figure(figsize=(12, 6))
sns.barplot(x='season', y='adjusted_offensive_efficiency', hue='tournament_winner', data=df)
plt.title('Adjusted Offensive Efficiency by Season and Tournament Winner Status')
plt.xlabel('Season')
plt.ylabel('Adjusted Offensive Efficiency')
plt.legend(title='Tournament Winner', loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create the same stacked bar plot but with adjusted_defensive_efficiency
plt.figure(figsize=(12, 6))
sns.barplot(x='season', y='adjusted_defensive_efficiency', hue='tournament_winner', data=df)
plt.title('Adjusted Defensive Efficiency by Season and Tournament Winner Status')
plt.xlabel('Season')
plt.ylabel('Adjusted Defensive Efficiency')
plt.legend(title='Tournament Winner', loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Graph correlation between adjusted_offensive_efficiency and adjusted_defensive_efficiency, colored by tournament_winner
plt.figure(figsize=(10, 6))
sns.scatterplot(x='adjusted_offensive_efficiency', y='adjusted_defensive_efficiency', hue='tournament_winner', data=df)
plt.title('Adjusted Offensive Efficiency vs Adjusted Defensive Efficiency by Tournament Winner Status')
plt.xlabel('Adjusted Offensive Efficiency')
plt.ylabel('Adjusted Defensive Efficiency')
plt.legend(title='Tournament Winner', loc='upper right')
plt.tight_layout()
plt.show()
```

# Distributions

```python
# Create boxplot comparing turnover_rate for tournament winners vs non-winners
plt.figure(figsize=(8, 6))
sns.boxplot(x='tournament_winner', y='topct_off', data=df)
plt.title('Turnover Rate by Tournament Winner Status')
plt.xlabel('Tournament Winner')
plt.ylabel('Offensive Turnover Rate')
plt.show()  
```

```python
# Explore outlier in turnover_rate for tournament losers
outliers = df[(df['tournament_winner'] == 0) & (df['topct_off'] > 50)]
print(outliers[['season', 'team_name', 'topct_off']])
```

```python
# Redo box chart with no values above 50 for turnover_rate
plt.figure(figsize=(8, 6))
sns.boxplot(x='tournament_winner', y='topct_off', data=df[df['topct_off'] <= 50])
plt.title('Turnover Rate by Tournament Winner Status (No Outliers)')
plt.xlabel('Tournament Winner')
plt.ylabel('Offensive Turnover Rate')
plt.show()
```

# Correlations

```python
# Make correlation heatmap for given variables
correlation_matrix = df[['adjusted_offensive_efficiency', 'adjusted_defensive_efficiency', 'tempo', 'off3ptfg', 'ftrate_off', 'experience', 'top_12_in_ap_top_25_during_week_6?', 'seed', 'tournament_winner']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap for Selected Variables')
plt.show()  
```

```python
# Plot scatterplot of seed vs tournament_winner, adding jitter to tournament_winner for better visualization and a regression line
#plt.figure(figsize=(10, 6))
#sns.scatterplot(x='seed', y='tournament_winner', data=df, jitter=True)
#sns.regplot(x='seed', y='tournament_winner', data=df, scatter=False, color='red')
#plt.title('Scatterplot of Seed vs Tournament Winner with Jitter and Regression Line')
#plt.xlabel('Seed')
#plt.ylabel('Tournament Winner')
#plt.show()

df['winner_jitter'] = df['tournament_winner'] + np.random.uniform(-0.05, 0.05, size=len(df))

sns.scatterplot(x='seed', y='winner_jitter', data=df, alpha=0.5)
plt.ylabel('Tournament Winner')
plt.yticks([0, 1], ['Loser', 'Winner'])
plt.show()
```

```python
# Calculate average seed for final four teams by season, visualize with line chart
final_four_teams = df[df['final_four?'] == 1]
avg_seed_final_four = final_four_teams.groupby('season')['seed'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(x='season', y='seed', data=avg_seed_final_four, marker='o')
plt.title('Average Seed for Final Four Teams by Season')
plt.xlabel('Season')
plt.ylabel('Average Seed')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()   
```
